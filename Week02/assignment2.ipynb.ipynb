```python
# I save the Label (Grouping) as a variable
TransitSelection = Transit['Label (Grouping)'].value_counts()
TransitSelection
```

```python
#I check what type (i.e string, value, series) of the TransitSelection column
type(TransitSelection)
```

```python
# I convert the series into a dataframe
TransitSelection = TransitSelection.reset_index()
TransitSelection
```

```python
#I check the columns name in the dataframe (previously a series)
TransitSelection.columns.to_list()
```

```python
# I change the name of the colunm names in the database
TransitSelection.columns = ['TransitType','Observation']
TransitSelection
```

```python
# I plot a simple line graph
TransitSelection.plot()
```

```python
#I have converted the line graph into a bar graph
TransitSelection.plot.bar(x = 'TransitType', y = 'Observation', title = 'Number of Vehicle Types')
```
